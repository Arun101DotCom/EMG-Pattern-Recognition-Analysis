```python
import pandas as pd  # For data handling
import numpy as np  # For numerical operations
from sklearn.model_selection import KFold  # For cross-validation
from sklearn.preprocessing import StandardScaler  # For scaling the data
from sklearn.neighbors import KNeighborsClassifier  # For KNN classifier
from sklearn.metrics import accuracy_score  # For model evaluation
import matplotlib.pyplot as plt  # For plotting
import time  # For measuring time

# List of data files with paths
files = {
    


    
    'MAV_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_mav_features.csv',
    'RMS_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_rms_features.csv',
    'VAR_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_var_features.csv',
    'WL_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_wl_features.csv',
    'ZC_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_zc_features.csv',
    'SSC_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_ssc_features.csv',
    'WAMP_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_wamp_features.csv',
    
    
    'MNF_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_mnf_features.csv',
    'MDF_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_mdf_features.csv',
    'PF_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_pf_features.csv',
    'PSD_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_psd_features.csv',
    'TP_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_tp_features.csv',
    
    
    
    'WT_Features': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\all_tf_energy_features.csv',
    
    
    'Combined_Time_domain': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\Combined_Time_Domain_Feature.csv',
    'Combined_Fequency_domain': r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\Combined_Frequency_Domain_Feature.csv',
    'combined_Time_Freq_& Tim-Freq_domain':r'C:\Users\c3054367\OneDrive - Newcastle University\lab\July\25072024\Combined\Matlab\Output_2507_150_75\Combined_Time_Freq_and_Time-Freq Domain_Feature.csv',
    
}


# Function to measure processing times for KNN on a given dataset with cross-validation
def measure_processing_times(file, label='Class'):
    timings = {'Load Time': 0, 'Scaler Time': 0, 'KNN Fit Time': 0, 'Total Time': 0}

    try:
        # Measure Load Time
        start_time = time.time()
        data = pd.read_csv(file)
        timings['Load Time'] = time.time() - start_time

        # Split the dataset into features and target variable
        X = data.drop(label, axis=1)
        Y = data[label]

        # Standardize the features
        start_time = time.time()
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        timings['Scaler Time'] = time.time() - start_time

        # Initialize the KNN classifier
        knn = KNeighborsClassifier(n_neighbors=5)  # Use default 5 neighbors

        # Perform cross-validation
        kf = KFold(n_splits=10, shuffle=True, random_state=42)
        start_time = time.time()
        
        knn_fit_total = 0
        accuracies = []
        for train_index, test_index in kf.split(X_scaled):
            X_train, X_test = X_scaled[train_index], X_scaled[test_index]
            Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
            
            # Fit the model
            start_time_fit = time.time()
            knn.fit(X_train, Y_train)
            knn_fit_total += time.time() - start_time_fit
            
            # Predict and evaluate
            Y_test_pred = knn.predict(X_test)
            accuracies.append(accuracy_score(Y_test, Y_test_pred))
        
        end_time = time.time()
        timings['KNN Fit Time'] = knn_fit_total
        timings['Total Time'] = timings['Load Time'] + timings['Scaler Time'] + timings['KNN Fit Time']

        # Print timing results
        print(f'\nTiming Results for {file}:')
        print(f'Load Time: {timings["Load Time"]:.2f} seconds')
        print(f'Scaler Time: {timings["Scaler Time"]:.2f} seconds')
        print(f'KNN Fit Time: {timings["KNN Fit Time"]:.2f} seconds')
        print(f'Total Time Without CV: {timings["Total Time"]:.2f} seconds\n')

        # Return the timings for plotting
        return timings

    except FileNotFoundError:
        print(f"File not found: {file}")
    except Exception as e:
        print(f"An error occurred while processing {file}: {e}")
        return None

# Measure processing times for each dataset and collect the results
results = {}

for label, file in files.items():
    timings = measure_processing_times(file)
    if timings is not None:
        results[label] = timings
    else:
        print(f"Skipping dataset: {label} due to errors")

# Extract labels for plotting
labels = list(results.keys())

# Plot processing time
fig, ax2 = plt.subplots(figsize=(15, 8))

load_times = [results[label]['Load Time'] for label in labels]
scaler_times = [results[label]['Scaler Time'] for label in labels]
knn_fit_times = [results[label]['KNN Fit Time'] for label in labels]
total_times = [results[label]['Total Time'] for label in labels]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

bars1 = ax2.bar(x - 1.5*width, load_times, width, label='Load Time', color='blue', edgecolor='black')
bars2 = ax2.bar(x - 0.5*width, scaler_times, width, label='Scaler Time', color='green', edgecolor='black')
bars3 = ax2.bar(x + 0.5*width, knn_fit_times, width, label='KNN Fit Time', color='purple', edgecolor='black')
bars4 = ax2.bar(x + 1.5*width, total_times, width, label='Total Time Without CV', color='red', edgecolor='black')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax2.set_xlabel('Datasets')
ax2.set_ylabel('Time Taken (seconds)')
ax2.set_title('Processing Time for Different Feature Datasets')
ax2.set_xticks(x)
ax2.set_xticklabels(labels, rotation=90)
ax2.legend()

fig.tight_layout()
plt.show()
```
